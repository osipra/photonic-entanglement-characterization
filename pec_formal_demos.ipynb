# Using `pec`: Formal Starter Demos

This notebook shows how to use the current `pec` package in a clean, formal way.

It focuses on the modules that exist now:
- `pec.states`
- `pec.metrics`

The goals are:
1. import the package from the repo,
2. inspect the available API,
3. build single-qubit and two-qubit states,
4. form projectors / density matrices,
5. compute basic state metrics,
6. evaluate Bell-state fidelities.


## 1. Setup

Run this notebook from the repository root. If needed, the next cell adds `src/` to the Python path.

```python
from pathlib import Path
import sys
import numpy as np

repo_root = Path.cwd()
src_path = repo_root / 'src'

if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

print('Repo root:', repo_root)
print('src path added:', src_path.exists(), src_path)
```

```python
import pec.states as states
import pec.metrics as metrics

print('Imported pec.states and pec.metrics successfully.')
```

## 2. Inspect the public API

This is a good first habit whenever you add a new module: inspect what the package actually exports.

```python
state_api = [name for name in dir(states) if not name.startswith('_')]
metric_api = [name for name in dir(metrics) if not name.startswith('_')]

print('states.py public names:')
for name in state_api:
    print('  ', name)

print('\nmetrics.py public names:')
for name in metric_api:
    print('  ', name)
```

## 3. Single-qubit states

The `states` module gives you the standard polarization basis states and related single-qubit states.

```python
H = states.ket_h()
V = states.ket_v()
D = states.ket_d()
A = states.ket_a()
R = states.ket_r()
L = states.ket_l()

print('|H> =', H)
print('|V> =', V)
print('|D> =', D)
print('|A> =', A)
print('|R> =', R)
print('|L> =', L)
```

```python
print('<H|H> =', np.vdot(H, H))
print('<V|V> =', np.vdot(V, V))
print('<H|V> =', np.vdot(H, V))
print('<D|A> =', np.vdot(D, A))
print('<R|L> =', np.vdot(R, L))
```

## 4. Projectors and density matrices

A state vector `|psi>` becomes a pure-state density matrix through the projector `|psi><psi|`.

```python
rho_H = states.projector(H)
rho_D = states.projector(D)

psi = (H + V)/ np.sqrt(2)
rho_psi = states.projector(psi)

print('rho_psi =')
print(rho_psi)

print('rho_H =')
print(rho_H)

print('\nrho_D =')
print(rho_D)
```

```python
print('Trace(rho_psi) =', np.trace(rho_psi))
print('Trace(rho_H) =', np.trace(rho_H))
print('Trace(rho_D) =', np.trace(rho_D))
print('Hermitian check for rho_D =', states.is_hermitian(rho_D))
```

## 5. Tensor-product states

Now we build two-qubit states from one-qubit polarization states.

```python
HV = states.tensor_ket('HV')
VH = states.tensor_ket('VH')
DD = states.tensor_ket('DD')

print('|HV> =', HV)
print('\n|VH> =', VH)
print('\n|DD> =', DD)
print('\nDimension of |HV> =', HV.shape)
```

```python
Pi_HV = states.tensor_projector('HV')
print('Projector |HV><HV| =')
print(Pi_HV)
```

## 6. Bell states

The Bell-state helpers are central to the entanglement-characterization part of the repo.

```python
bells = states.bell_states()
for label, ket in bells.items():
    print(label)
    print(ket)
    print('norm =', np.vdot(ket, ket))
    print()
```

```python
phi_plus = states.bell_state('phi_plus')
rho_phi_plus = states.projector(phi_plus)

print('|Phi+> =')
print(phi_plus)

print('\nrho_{Phi+} =')
print(rho_phi_plus)
```

## 7. Basic metrics

This section uses `pec.metrics` on pure and mixed states.

```python
print('purity(|H><H|) =', metrics.purity(rho_H))
print('purity(|Phi+><Phi+|) =', metrics.purity(rho_phi_plus))
```

```python
fidelity_general_name = None
for candidate in ['fidelity', 'fidelity_uhlmann', 'density_matrix_fidelity', 'state_fidelity']:
    if hasattr(metrics, candidate):
        fidelity_general_name = candidate
        break

print('General density-matrix fidelity function found:', fidelity_general_name)
fidelity_general = getattr(metrics, fidelity_general_name) if fidelity_general_name is not None else None
```

```python
print('F(|H>, rho_H) =', metrics.fidelity_pure(rho_H, H))
print('F(|D>, rho_H) =', metrics.fidelity_pure(rho_H, D))

if fidelity_general is not None:
    print('F(rho_H, rho_H) =', fidelity_general(rho_H, rho_H))
    print('F(rho_phi_plus, rho_phi_plus) =', fidelity_general(rho_phi_plus, rho_phi_plus))
```

## 8. Bell-state fidelities

A natural demo for this repo is: given a two-qubit density matrix, how much weight does it have on each Bell state?

```python
bell_fids_ideal = metrics.bell_state_fidelities(rho_phi_plus)
print('Bell-state fidelities for ideal |Phi+><Phi+|:')
for label, value in bell_fids_ideal.items():
    print(f'  {label}: {value:.6f}')
```

```python
# A simple mixed state: 70% |Phi+><Phi+| + 30% |HV><HV|
rho_mixed = 0.7 * rho_phi_plus + 0.3 * states.tensor_projector('HV')
rho_mixed = states.make_hermitian(rho_mixed)
rho_mixed = states.trace_normalize(rho_mixed)

print('Trace(rho_mixed) =', np.trace(rho_mixed))
print('Purity(rho_mixed) =', metrics.purity(rho_mixed))

bell_fids_mixed = metrics.bell_state_fidelities(rho_mixed)
print('\nBell-state fidelities for mixed state:')
for label, value in bell_fids_mixed.items():
    print(f'  {label}: {value:.6f}')
```

## 9. Numerical cleanup demo

This shows why helpers like `make_hermitian` and `trace_normalize` exist.

```python
bad_rho = rho_phi_plus.copy().astype(complex)
bad_rho[0, 1] += 1e-6 + 2e-6j
bad_rho[2, 2] *= 1.05

print('Before cleanup:')
print('  Hermitian?', states.is_hermitian(bad_rho))
print('  Trace =', np.trace(bad_rho))

rho_fixed = states.make_hermitian(bad_rho)
rho_fixed = states.trace_normalize(rho_fixed)

print('\nAfter cleanup:')
print('  Hermitian?', states.is_hermitian(rho_fixed))
print('  Trace =', np.trace(rho_fixed))
```
